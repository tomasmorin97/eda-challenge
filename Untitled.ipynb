# Leer los CSV

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_diamonds_city = pd.read_csv("./data/diamonds_city.csv")
df_diamonds_city.head()
```

```python
df_diamonds_clarity = pd.read_csv("./data/diamonds_clarity.csv")
df_diamonds_clarity
```

```python
df_diamonds_color = pd.read_csv("./data/diamonds_color.csv")
df_diamonds_color.head()
```

```python
df_diamonds_cut = pd.read_csv("./data/diamonds_cut.csv")
df_diamonds_cut
```

```python
df_diamonds_dimensions = pd.read_csv("./data/diamonds_dimensions.csv")
df_diamonds_dimensions.info()
```

```python
df_diamonds_properties = pd.read_csv("./data/diamonds_properties.csv")
df_diamonds_properties.info()
```

```python
df_diamonds_transactional = pd.read_csv("./data/diamonds_transactional.csv")
df_diamonds_transactional.info()
```

```python
df1 = df_diamonds_dimensions.merge(df_diamonds_properties, on="index_id", how="left")
df1
```

```python
df_diamonds_id = df1.merge(df_diamonds_transactional, on="index_id", how="left")
df_diamonds_id
```

```python
df_diamonds1 = df_diamonds_id.merge(df_diamonds_city, on='city_id', how='left')
df_diamonds2 = df_diamonds1.merge(df_diamonds_clarity, on='clarity_id', how='left')
df_diamonds3 = df_diamonds2.merge(df_diamonds_color, on='color_id', how='left')
df_diamonds4 = df_diamonds3.merge(df_diamonds_cut, on='cut_id', how='left')
df_diamonds = df_diamonds4[['index_id','depth','table','x','y','z','price','carat','city','clarity','color','cut']]
df_diamonds
```

```python
fig, axs = plt.subplots(ncols = 4,
                       nrows = 1,
                       figsize= (11,3))
axs[0].boxplot(df_diamonds['table'])
axs[0].set(title='Table')
axs[1].boxplot(df_diamonds['depth'])
axs[1].set(title='Depth')
axs[2].boxplot(df_diamonds['price'])
axs[2].set(title='Price')
axs[3].boxplot(df_diamonds['carat'])
axs[3].set(title='Carat')
```

```python

# Muestra el gráfico
sns.heatmap(df_diamonds[['table','depth','price','carat','x','y','z']].corr(), vmin=-1, vmax=+1, annot=True, cmap="coolwarm")
```
